# Step 2 : Cleaning the Data

```python
retail.isnull().sum() #isnull() function detect missing values in the given series object
```

```python
#The drop() function is used to drop specified labels from rows or columns

supermarket = retail.drop(['Country','Region','Postal Code'],axis=1) 
```

```python
supermarket.head()
```

# Step 3 : Relationship Analysis

```python
#Correlation Matrix - corr() is used to create the correlation matrix
#You'll have to make sure that all the elements in the matrix are of numeric type

corelation = supermarket.corr()
```

```python
#heatmaps are typically used to visualize correlation matrices

sb.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True)
```

```python
# Pair plots are a great method to identify trends for follow-up analysis

sb.pairplot(supermarket)
```

```python
#relplot(), this is a figure-level function for visualizing statistical relationships using two common approaches

sb.relplot(data=supermarket, x='Profit', y='Discount', hue='Sales')
```

```python
sb.relplot(data=supermarket, x='Sales', y='Quantity', hue='Discount')
```

```python
#Categorical Plot
#stripplot creates scatter plot based on the category
sb.stripplot(x='Category', y='Sales', data=supermarket)
```

```python
sb.countplot(data=supermarket, y='Sub-Category')
```
